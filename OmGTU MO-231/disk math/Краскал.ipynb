```
static private void DFS(int v,  HashSet<int> worked, List<int[]> E)
{
    worked.Add(v);
    foreach (int[] sravnenie in E)
    {
        if (sravnenie[0] == v && !worked.Contains(sravnenie[1]))
        {
            DFS(sravnenie[1], worked, E);
        }
        else if (sravnenie[1] == v && !worked.Contains(sravnenie[0]))
        {
            DFS(sravnenie[0], worked, E);
        }
    }
}

static public bool IsIn(int v, int u, List<int[]> E, HashSet<int> V) //находятся ли 2 вершины в одной компоненте связанности
{
    HashSet<int> worked = new HashSet<int>();

    DFS(v, worked, E);

    return worked.Contains(u);
}
```

```
int V = 6;

List<int[]> E = new List<int[]>(new int[][]
    {
        new int[] {1, 2, 6},
        new int[] {1, 3, 7},
        new int[] {1, 6, 5},
        new int[] {2, 6, 3},
        new int[] {2, 3, 3},
        new int[] {2, 4, 4},
        new int[] {2, 5, 4},
        new int[] {3, 5, 5},
        new int[] {3, 4, 6},
        new int[] {3, 6, 3},
        new int[] {4, 5, 5},
    }
);

List<int[]> SortE = E.OrderBy(arr => arr[2]).ToList();
HashSet<int> newV = new HashSet<int>();
List<int[]> newE = new List<int[]>();
int i = 0;
while (newV.Count != V)
{
    int[] v = SortE[i];
    if (!newV.Contains(v[0]) || !newV.Contains(v[0]) || !IsIn(v[0],v[1], newE, newV))
    {
        newV.Add(v[0]);
        newV.Add(v[1]);
        newE.Add(v);
    }
    i++;
}
newE
```
